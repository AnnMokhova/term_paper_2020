```python
import pymorphy2
from collections import Counter
```

```python
periods = ['Tosia_hears_1.txt', 'Tosia_hears_2.txt', 'Tosia_hears_3.txt',
          'Tosia_hears_4.txt', 'Tosia_says_1.txt', 'Tosia_says_2.txt',
          'Tosia_says_3.txt', 'Tosia_says_4.txt', 'Yasha_hears_1.txt',
          'Yasha_hears_2.txt', 'Yasha_hears_3.txt', 'Yasha_says_1.txt',
          'Yasha_says_2.txt', 'Yasha_says_3.txt']
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
def NounsAnalysis(period):
    cases = []
    numbers = []
    genders = []
    with open(period, 'r+', encoding='utf-8') as f:
        text = f.readlines()
        summ = len(text)
        for word in text:
            word = word[:-1]
            p = morph.parse(word)[0]
            cases.append(p.tag.case)
            numbers.append(p.tag.number)
            genders.append(p.tag.gender)
    print(summ)
    print(Counter(cases))
    print(Counter(numbers))
    print(Counter(genders))
```

```python
for period in periods:
    print(period)
    NounsAnalysis(period)
    print('\n')
```

## Статистика значений категории рода и одушевленности без имен детей

```python
def NounsAnalysis2(period):
    genders = []
    anims = []
    with open(period, 'r+', encoding='utf-8') as f:
        text = f.readlines()
        for word in text:
            word = word[:-1]
            p = morph.parse(word)[0]
            if p.normal_form != 'яша' and p.normal_form != 'тося':
                genders.append(p.tag.gender)
                anims.append(p.tag.animacy)
    print(len(genders))
    print(Counter(genders))
    print(Counter(anims))
```

```python
for period in periods:
    print(period)
    NounsAnalysis2(period)
    print('\n')
```
